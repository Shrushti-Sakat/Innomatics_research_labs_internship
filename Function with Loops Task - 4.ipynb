## Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return False, "Recharge amount must be at least Rs.50."

    if amount not in valid_plans:
        return False, f"Invalid plan. Choose from: {valid_plans}"

    return True, f"Recharge of Rs.{amount} is valid. Processing..."


def mobile_recharge_system():
    while True:
        try:
            amount = int(input("Enter recharge amount (Rs.): "))
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
            continue

        is_valid, message = validate_recharge(amount)
        print(message)

        if is_valid:
            break

        retry = input("Do you want to try again? (yes/no): ").strip().lower()
        if retry != "yes":
            print("Recharge cancelled.")
            break


mobile_recharge_system()
```

## Problem 2: Inventory Reorder Alert System

```python
def check_inventory(inventory):
    for product, stock in inventory.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock: {stock})")
        else:
            print(f"{product}: Stock OK (Stock: {stock})")


inventory = {
    "Pen": 10,
    "Notebook": 25,
    "Eraser": 5,
    "Stapler": 18,
    "Marker": 3
}

check_inventory(inventory)
```

## Problem 3: Student Result Processing System

```python
def process_student_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)
    print(f"Total Marks   : {total}")
    print(f"Average Marks : {average:.2f}")

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


student_marks = [78, 45, 88, 32, 90]
result = process_student_result(student_marks)
print(f"Result        : {result}")
```

## Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, is_peak_hour):
    base_fare = 50
    rate_per_km = 12

    fare = base_fare + (distance * rate_per_km)

    if is_peak_hour:
        fare = fare * 1.25

    return fare


def cab_fare_estimator():
    while True:
        try:
            distance = float(input("Enter distance in km: "))
            if distance <= 0:
                print("Distance must be greater than 0.")
                continue
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
            continue

        peak_input = input("Is it peak hour? (yes/no): ").strip().lower()
        is_peak = peak_input == "yes"

        fare = calculate_fare(distance, is_peak)
        print(f"Estimated Fare: Rs.{fare:.2f}")

        if is_peak:
            print("Note: Peak hour surcharge of 25 percent applied.")

        retry = input("Do you want to estimate another fare? (yes/no): ").strip().lower()
        if retry != "yes":
            print("Thank you for using the Cab Fare Estimator.")
            break


cab_fare_estimator()
```

## Problem 5: Employee Attendance Eligibility Checker

```python
def check_attendance_eligibility(attendance):
    total_days = len(attendance)
    present_days = 0

    for record in attendance:
        if record == "P":
            present_days += 1

    attendance_percentage = (present_days / total_days) * 100
    print(f"Total Days    : {total_days}")
    print(f"Days Present  : {present_days}")
    print(f"Attendance    : {attendance_percentage:.2f}%")

    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_record = ["P", "A", "P", "P", "P", "A", "P", "P", "A", "P"]
status = check_attendance_eligibility(attendance_record)
print(f"Eligibility   : {status}")
```

## Problem 6: Password Strength Checker

```python
def check_password_strength(password):
    special_characters = ['@', '#', '$']
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak", "Password must be at least 8 characters long."

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_characters:
            has_special = True

    if not has_digit:
        return "Weak", "Password must contain at least one digit."

    if not has_special:
        return "Weak", "Password must contain at least one special character (@, #, $)."

    return "Strong", "Password meets all the required criteria."


test_passwords = [
    "hello",
    "hello1234",
    "hello@world",
    "hello@1234"
]

for pwd in test_passwords:
    strength, message = check_password_strength(pwd)
    print(f"Password: {pwd}")
    print(f"Strength: {strength}")
    print(f"Message : {message}")
    print("-" * 40)
```
