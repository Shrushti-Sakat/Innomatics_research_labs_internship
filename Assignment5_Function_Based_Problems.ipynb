## Q1. Social Media — Post Engagement Analyzer

```python
# I will store individual like counts in a list and add them up inside the function

def post_engagement_checker(daily_likes):
    count = 0
    for each_day in daily_likes:      # going through each day's likes one by one
        count = count + each_day

    if count >= 1000:
        result = "Viral Post"
    else:
        result = "Normal Engagement"

    print("Total Likes:", count)
    print("Post Status:", result)


# Suppose a post got these likes over 5 days
post_likes = [320, 215, 180, 205, 130]
post_engagement_checker(post_likes)
```

## Q3. Agriculture — Rainfall Adequacy Checker

```python
# The function takes a list of monthly readings and the required minimum level

def rainfall_season_check(monthly_readings, min_required):
    total = 0
    for reading in monthly_readings:    # adding up all monthly values
        total += reading

    avg = total / len(monthly_readings)  # calculating the mean

    if avg >= min_required:
        season_result = "Adequate Rainfall"
    else:
        season_result = "Inadequate Rainfall"

    print("Average Rainfall:", int(avg))
    print("Rainfall Status:", season_result)


# Rainfall data for 6 months (in mm) for a farm in Nashik
nashik_rainfall = [55, 90, 78, 62, 84, 71]
needed_level = 70

rainfall_season_check(nashik_rainfall, needed_level)
```

## Q4. Social Media — Duplicate Account Detection

```python
# Using a set to catch repeated usernames — sets automatically remove duplicates

def find_duplicate_users(user_list):
    unique_set = set(user_list)

    if len(user_list) != len(unique_set):   # mismatch means there are repeats
        print("Duplicate Accounts Found: Yes")

        # Now I want to find exactly which usernames are repeated
        tracker = set()
        repeated = []
        for name in user_list:
            if name in tracker and name not in repeated:
                repeated.append(name)
            tracker.add(name)

        print("Flagged Accounts:", repeated)
    else:
        print("Duplicate Accounts Found: No")
        print("All usernames are unique.")


# Sample list with some repeat usernames
registered_users = ["dev_kartik", "sneha_21", "rohan_k", "dev_kartik", "pooja_r", "sneha_21"]
find_duplicate_users(registered_users)
```

## Q6. Agriculture — Premium Crop Price Filter

```python
# I'll loop through all crop prices and split them into two groups

def separate_premium_crops(price_list, cutoff=2000):
    premium = []
    regular = []

    for price in price_list:
        if price > cutoff:
            premium.append(price)    # qualifies as premium
        else:
            regular.append(price)    # goes into regular category

    print("Premium Crops:", premium)
    print("Regular Crops:", regular)


# Prices of different crops at the mandi (in Rs. per quintal)
mandi_prices = [1200, 2500, 950, 3100, 2050, 1800, 2800]
separate_premium_crops(mandi_prices)
```

## Q10. Smart Electricity Bill Analyzer

```python
# The billing logic builds up in layers — first slab, then second, then third

def electricity_bill_calculator(consumer, units):
    bill = 0

    if units <= 100:
        bill = units * 3

    elif units <= 200:
        # first 100 units at Rs.3, remaining at Rs.5
        bill = (100 * 3) + ((units - 100) * 5)

    else:
        # first 100 at Rs.3, next 100 at Rs.5, rest at Rs.7
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Classify based on final bill amount
    if bill < 500:
        category = "Low Usage"
    elif bill <= 1500:
        category = "Moderate Usage"
    else:
        category = "High Usage"

    print("Consumer      :", consumer)
    print("Units Used    :", units)
    print("Total Bill    : Rs.", bill)
    print("Usage Level   :", category)

    return bill, category


# Three households with different consumption levels
print("=== Household 1 ===")
electricity_bill_calculator("Suresh Pawar", 75)

print()
print("=== Household 2 ===")
electricity_bill_calculator("Geeta Shinde", 160)

print()
print("=== Household 3 ===")
electricity_bill_calculator("Amit Thakur", 310)
```
