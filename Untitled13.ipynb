```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="distilgpt2",
    max_new_tokens=80,
    do_sample=False,
    repetition_penalty=2.0
)

from langchain_community.llms import HuggingFacePipeline
llm = HuggingFacePipeline(pipeline=pipe)
```

## 🔍 What is LangChain?

LangChain is an open-source framework that simplifies building applications using LLMs.

### Key Components:
- LLMs (Language Models)
- Prompt Templates
- Chains
- Memory

It is widely used in chatbots, automation, and AI assistants.

```python
prompt = PromptTemplate.from_template(
    "LangChain is a framework used in AI."
)

chain = prompt | llm

response = chain.invoke({})

text = response.strip()

# Split into sentences
sentences = text.split(".")

# Clean valid sentences only
clean_points = []
for s in sentences:
    s = s.strip()
    if len(s) > 20 and "LangChain" in s:
        clean_points.append(s)
    if len(clean_points) == 3:
        break

# If model fails, fallback manual answer
if len(clean_points) < 3:
    clean_points = [
        "LangChain is a framework for building applications using language models",
        "It helps connect prompts, memory, and tools",
        "It is widely used in chatbots and AI systems"
    ]

# Print final output
for i, point in enumerate(clean_points, 1):
    print(f"{i}. {point}")
```

## 🧩 Prompt Templates

They help create reusable and flexible prompts.

## 🔗 Chains in LangChain

Chains combine prompts and models:

prompt | llm

This creates a simple pipeline for execution.

```python
prompt = PromptTemplate.from_template(
    "Artificial Intelligence is an important technology."
)

chain = prompt | llm

response = chain.invoke({})

print(response)
```
