1. User Login Check

```python
username="admin"
password="1234"
input1=input("Enter username: ")
input2=input("Enter Password: ")
if input1==username and input2==password:
    print("Login successful")
else:
    print("Invalid credentials")
```

2.Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
totalpassed=0
totalfailed=0
for m in marks:
    if m>=50:
        totalpassed += 1
    else:
        totalfailed += 1
print("Total passed:", totalpassed)
print("Total failed:", totalfailed)
```

3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_list=[]
for i in names:
    x=i.strip().lower()
    cleaned_list.append(x)
print(cleaned_list)
```


4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    if len(msg) > 10:
        print("Long Message:",msg)
    else:
        print(msg,", len(msg):",len(msg))
```

5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for log in logs:
    if log == "ERROR":
        count+=1
print("Total error count:",count)
```
