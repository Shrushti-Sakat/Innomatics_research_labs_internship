```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"
    
    def turn_on(self):
        self.status = "ON"
    
    def turn_off(self):
        self.status = "OFF"
    
    def display_status(self):
        print(f"{self.light_name} is {self.status}")

light_name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ")

light = SmartLight(light_name)

if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()
else:
    print("Invalid Action")

light.display_status()
```

```python
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display(self):
        print("\nEmployee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.dept}")
    
name = input("\nEmployee Name:")
emp_id = input("Employee ID:")
dept = input("Department:")
emp=Employee(name,emp_id,dept)
emp.display()
```
